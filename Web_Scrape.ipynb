```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
```

```python
headers = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/133.0.0.0 Safari/537.36'}
```

```python
district = ['kathmandu/','lalitpur/']
page = '?page='
```

```python
right_dist = {}
left_dist = {}
Price_Rs  = []
District = []
VDC = []
Direction = []
Bedroom = []
Bathroom = []
Living = []
Kitchen = []
Total = []
Parking = []
Built = []
Land_area = []
Road_size  = []
href = []
rightt = {
    'District': District,
    'VDC': VDC,
    'Direction': Direction,
    'Bedroom': Bedroom,
    'Bathroom': Bathroom,
    'Living': Living,
    'Kitchen': Kitchen,
    'Total': Total,
    'Parking': Parking,
    'Built': Built
}
leftt = {
    'Road_size': Road_size,
    'Land_area': Land_area,
    'Price_Rs': Price_Rs
}
```

```python
for j in district:
    for i in range(1,16):
        url = 'https://www.gharghaderi.com/house-for-sale/' + j
        url = url + page + str(i)
        html_text = requests.get(url,headers = headers)
        soup = BeautifulSoup(html_text.text,'lxml')
        link = soup.find('div',class_="grid-thirds2 grid-thirds" )
        for k in link.find_all('a'):
            href.append(k.get('href'))
```

```python
for i in href:
    url = 'https://www.gharghaderi.com/house/'+i
    html_text = requests.get(url,headers = headers)
    soup = BeautifulSoup(html_text.text,'lxml')
    right = soup.find_all("div",class_="right")
    for i in right:
        list = i.find_all('tr')
        for j in list:
            key = j.find('td',class_ = 'test').text
            pair = j.find('td',class_ = 'test1').text
            right_dist[key.strip().split()[0]] = pair.strip()
        for key, value in right_dist.items():
            if key in rightt:
                try:
                    rightt[key].append(value)
                except:
                    pass
    left = soup.find_all("div",class_="left")
    for i in left:
        list = i.find_all('tr')
        for j in list:
            key = j.find('td',class_ = 'test').text
            pair = j.find('td',class_ = 'test1').text
            left_dist[key.lstrip().rstrip().replace(':','').rstrip().replace(' ','_')] = pair.strip()
        for key, value in left_dist.items():
            if key in leftt:
                try:
                    leftt[key].append(value)
                except:
                    pass 
```

```python
df = pd.DataFrame({'District': District,
    'VDC': VDC,
    'Direction': Direction,
    'Bedroom': Bedroom,
    'Bathroom': Bathroom,
    'Living': Living,
    'Kitchen': Kitchen,
    'Total': Total,
    'Parking': Parking,
    'Built': Built,
    'Road_size': Road_size,
    'Land_area': Land_area,
    'Price_Rs': Price_Rs})
```

```python
def empty(column):
    df[column] = df[column].apply(lambda x: np.nan if x == '' else x)
```
